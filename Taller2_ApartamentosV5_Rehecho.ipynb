### 2. Entrenamiento del modelo de Machine Learning

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, FunctionTransformer
from sklearn.impute import SimpleImputer

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
```

```python
# Separamos el set en train, test, val

y_raw = df[TARGET].astype(float)
X = df.drop(columns=[TARGET], errors="ignore")

num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X.select_dtypes(exclude=[np.number]).columns.tolist()

X_train_full, X_temp, y_train_full, y_temp = train_test_split(
    X, y_raw, test_size=0.40, random_state=42
)
X_test, X_val, y_test, y_val = train_test_split(
    X_temp, y_temp, test_size=0.50, random_state=42
)
```

```python
# Verificamos multicolinealidad

X_num_train = X_train_full[num_cols].copy()

imp_num_tmp = SimpleImputer(strategy="median", keep_empty_features=True)
X_num_imp = pd.DataFrame(
    imp_num_tmp.fit_transform(X_num_train),
    columns=X_num_train.columns,
    index=X_num_train.index
)
```

```python
vars_principales = [
    "area", "habitaciones", "estrato", "parqueaderos"
]
```

```python
vars_prescindibles = [
    c for c in X_num_imp.columns
    if c not in vars_principales
]
```

```python
def calcular_vif(df):
    vif = pd.DataFrame()
    vif["variable"] = df.columns
    vif["VIF"] = [
        variance_inflation_factor(df.values, i)
        for i in range(df.shape[1])
    ]
    return vif

X_vif = X_num_imp.copy()
cols_to_drop_vif = []

while True:
    vif_df = calcular_vif(X_vif)
    max_vif = vif_df["VIF"].max()
    var_max = vif_df.loc[vif_df["VIF"].idxmax(), "variable"]

    if max_vif > 10 and var_max not in vars_principales:
        X_vif = X_vif.drop(columns=[var_max])
        cols_to_drop_vif.append(var_max)
    else:
        break
```

```python
cols_to_drop_vif
```

```python
num_cols_linear = [c for c in num_cols if c not in cols_to_drop_vif]

print("Num. FULL:", len(num_cols), "| Num. LINEAR:", len(num_cols_linear))
```

```python
#Preprocesamiento

# Común para categóricas (igual para todos los modelos)
cat_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))  # sparse por defecto
])

# Para modelos LINEALES: imputación + escalado en numéricas reducidas
num_linear = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

preprocess_linear = ColumnTransformer(
    transformers=[
        ("num", num_linear, num_cols_linear),
        ("cat", cat_transformer, cat_cols),
    ],
    remainder="drop"
)


# Para ÁRBOLES: imputación sin escalado con TODAS las numéricas
num_tree = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
])

preprocess_trees = ColumnTransformer(
    transformers=[
        ("num", num_tree, num_cols),
        ("cat", cat_transformer, cat_cols),
    ],
    remainder="drop"
)

preprocess_ridge = ColumnTransformer(
    transformers=[
        ("num", num_linear, num_cols),
        ("cat", cat_transformer, cat_cols),
    ],
    remainder="drop"
)

def log1p_transform(y):
    return np.log1p(y)

def inv_log1p(yhat):
    return np.expm1(yhat)

y_train_full_log = log1p_transform(y_train_full)
y_test_log = log1p_transform(y_test)
y_val_log = log1p_transform(y_val)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def regression_report(y_true, y_pred, label):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)      # no 'squared' kwarg
    rmse = np.sqrt(mse)                           # compute RMSE manually
    r2 = r2_score(y_true, y_pred)

    within_10 = float(np.mean(np.abs(y_pred - y_true) / y_true <= 0.10))

    print(f"\n--- {label} ---")
    print(f"MAE:  {mae:,.0f} COP")
    print(f"RMSE: {rmse:,.0f} COP")
    print(f"R²:   {r2:0.4f}")
    print(f"% dentro de ±10%: {within_10*100:0.2f}%")

    return {"MAE": mae, "RMSE": rmse, "R2": r2, "ACC±10%": within_10}
```

```python
from sklearn.feature_selection import VarianceThreshold

linreg_pipe = Pipeline(steps=[
    ("prep", preprocess_linear),
    ("vt", VarianceThreshold(threshold=0.0)),    # elimina columnas con varianza 0
    ("model", LinearRegression())
])

try:
    linreg_pipe.fit(X_train_full, y_train_full_log)
    y_test_pred_log = linreg_pipe.predict(X_test)

except Exception as e:
    print("LinearRegression failed with:", repr(e))
    print("Falling back to numerically-stable Ridge(alpha=1e-6)...")

    linreg_pipe = Pipeline(steps=[
        ("prep", preprocess_linear),
        ("vt", VarianceThreshold(threshold=0.0)),
        ("model", Ridge(alpha=1e-6, random_state=42))  # casi equivalente a OLS pero estable
    ])

    linreg_pipe.fit(X_train_full, y_train_full_log)
    y_test_pred_log = linreg_pipe.predict(X_test)

# Back-transform to COP
y_test_pred = inv_log1p(y_test_pred_log)

# Evaluate
metrics_baseline = regression_report(y_test, y_test_pred, "Linear baseline (TEST)")

print("\nBaseline LR metrics:")
metrics_baseline
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

# Pipeline individual para Ridge
ridge_pipe = Pipeline(steps=[
    ("prep", preprocess_ridge),
    ("model", Ridge(random_state=42))
])

ridge_grid = {
    "model__alpha": [0.1, 1.0, 10.0, 100.0]
}

print("\n>>> GridSearch: RIDGE")
ridge_gs = GridSearchCV(
    ridge_pipe,
    ridge_grid,
    cv=3,
    scoring="neg_mean_absolute_error",
    n_jobs=-1,
    verbose=2
)

ridge_gs.fit(X_train_full, y_train_full_log)
y_pred_log = ridge_gs.best_estimator_.predict(X_test)
y_pred = inv_log1p(y_pred_log)

print("Mejores parámetros RIDGE:", ridge_gs.best_params_)
results_ridge = regression_report(y_test, y_pred, "RIDGE (TEST)")
results_ridge
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_pipe = Pipeline(steps=[
    ("prep", preprocess_trees),
    ("model", RandomForestRegressor(random_state=42, n_jobs=-1))
])

rf_grid = {
    "model__n_estimators": [600],
    "model__max_depth": [None],
    "model__min_samples_leaf": [1]
}

print("\n>>> GridSearch: RF")
rf_gs = GridSearchCV(
    rf_pipe,
    rf_grid,
    cv=3,
    scoring="neg_mean_absolute_error",
    n_jobs=-1,
    verbose=2
)

rf_gs.fit(X_train_full, y_train_full_log)

y_pred_log = rf_gs.best_estimator_.predict(X_test)
y_pred = inv_log1p(y_pred_log)

print("Mejores parámetros RF:", rf_gs.best_params_)
results_rf = regression_report(y_test, y_pred, "RF (TEST)")
results_rf
```

### 3. Análisis cuantitativo de resultados del modelo

```python
y_val_real = inv_log1p(y_val_log)

modelos = [
    ("RandomForest", rf_gs.best_estimator_),
    ("Ridge",        ridge_gs.best_estimator_),
    ("Linear",       linreg_pipe),
]

resultados = []
preds_dict = {}

for nombre, est in modelos:
    y_pred_log = est.predict(X_val)
    y_pred = inv_log1p(y_pred_log)
    preds_dict[nombre] = y_pred

    mae  = np.mean(np.abs(y_val_real - y_pred))
    rmse = np.sqrt(np.mean((y_val_real - y_pred) ** 2))

    
    ss_res = np.sum((y_val_real - y_pred) ** 2)
    ss_tot = np.sum((y_val_real - np.mean(y_val_real)) ** 2)
    r2 = 1 - ss_res/ss_tot if ss_tot > 0 else np.nan

    
    pct10 = float(
        np.mean(
            np.abs(y_pred - y_val_real) / np.maximum(y_val_real, 1e-12) <= 0.10
        )
    )

    resultados.append({
        "Modelo": nombre,
        "MAE": mae,
        "RMSE": rmse,
        "R²": r2,
        "% dentro ±10%": pct10
    })


df_metrics_val = pd.DataFrame(resultados).sort_values("MAE").reset_index(drop=True)



df_show = df_metrics_val.copy()

# MAE y RMSE en COP normal (con miles)
df_show["MAE"]  = df_show["MAE"].apply(lambda x: f"{x:,.0f}")
df_show["RMSE"] = df_show["RMSE"].apply(lambda x: f"{x:,.0f}")

# R² con 3 decimales
df_show["R²"] = df_show["R²"].apply(lambda x: f"{x:.3f}")

# porcentaje con formato %
df_show["% dentro ±10%"] = df_show["% dentro ±10%"].apply(lambda x: f"{x:.2%}")

display(df_show)



from matplotlib.ticker import FuncFormatter
mill_formatter = FuncFormatter(lambda x, pos: f"{x/1_000_000:,.0f}")


#--------------------
order = np.argsort(np.array(y_val_real))
idx = np.arange(len(order))
y_sorted = np.array(y_val_real)[order]

for nombre, y_pred in preds_dict.items():
    y_pred_sorted = np.array(y_pred)[order]

    plt.figure(figsize=(12, 5))

    
    ymin = np.minimum(y_sorted, y_pred_sorted)
    ymax = np.maximum(y_sorted, y_pred_sorted)
    plt.vlines(idx, ymin, ymax, alpha=0.18, linewidth=1)

   
    plt.scatter(
        idx,
        y_pred_sorted,
        s=20,
        facecolors='none',
        edgecolors='C1',
        label=f"Predicho ({nombre})",
        zorder=2
    )

  
    plt.scatter(
        idx,
        y_sorted,
        s=22,
        color='C0',
        alpha=0.85,
        label="Real",
        zorder=4  
    )

    
    plt.title("Comparación de curvas en validación")
    plt.xlabel("Observaciones (real y predicha)")
    plt.ylabel("y (millones de COP)")
    ax = plt.gca()
    ax.yaxis.set_major_formatter(mill_formatter)
    plt.legend()
    plt.tight_layout()
    plt.show()

#---------------------

plot_df = pd.DataFrame({"Real": y_val_real})
for nombre, y_pred in preds_dict.items():
    plot_df[nombre] = y_pred

plt.figure(figsize=(8, 5))
plt.boxplot([plot_df[col] for col in plot_df.columns], labels=plot_df.columns)
plt.title("Distribución: y Real vs y Predicha (VALIDATION)")
plt.ylabel("y (millones de COP)")
ax = plt.gca()
ax.yaxis.set_major_formatter(mill_formatter)
plt.tight_layout()
plt.show()
```

### 4. Análisis cualitativo de resultados del modelo

```python
import shap, numpy as np, pandas as pd, scipy.sparse as sp
shap.initjs()

rf_pipe_best = rf_gs.best_estimator_
prep_rf      = rf_pipe_best.named_steps["prep"]
rf_model     = rf_pipe_best.named_steps["model"]

Xtr_rf = prep_rf.transform(X_train_full)
Xte_rf = prep_rf.transform(X_test)

bg_size = min(300, Xtr_rf.shape[0])

if sp.issparse(Xtr_rf):
    Xtr_bg = Xtr_rf[:bg_size].toarray()
else:
    Xtr_bg = Xtr_rf[:bg_size]

if sp.issparse(Xte_rf):
    Xte_dense = Xte_rf.toarray()
else:
    Xte_dense = Xte_rf

try:
    feat_names_rf = prep_rf.get_feature_names_out()
except:
    feat_names_rf = np.array([f"f{i}" for i in range(Xte_dense.shape[1])])

explainer_rf = shap.TreeExplainer(rf_model, data=Xtr_bg)

shap_values_rf = explainer_rf(Xte_dense, check_additivity=False)

X_display_rf = pd.DataFrame(Xte_dense, columns=feat_names_rf)
shap_values_rf.data = X_display_rf.values
shap_values_rf.feature_names = list(feat_names_rf)


shap.plots.beeswarm(shap_values_rf, max_display=20)


shap.force_plot(
    explainer_rf.expected_value,
    shap_values_rf.values[:100,:],
    X_display_rf.iloc[:100,:],
    feature_names=feat_names_rf
)


sample_index = 160
print("Valor real (log→COP):", np.expm1(y_test.to_numpy()[sample_index]))

pred_tree_log = rf_pipe_best.predict(X_test)
pred_tree     = np.expm1(pred_tree_log)
print("Valor predicho:", pred_tree[sample_index])

shap.plots.waterfall(shap_values_rf[sample_index])
```
